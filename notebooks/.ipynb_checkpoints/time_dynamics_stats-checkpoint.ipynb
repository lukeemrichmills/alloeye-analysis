```python

def cochrane_orcutt(X, y, rho, n_it=100, ax=None, c='m'):
    """
    Implement the Cochrane-Orcutt procedure.
    
    Parameters:
    y -- a numpy array of shape (n, ) (dependent variable)
    X -- a numpy array of shape (n, p) (independent variables)
    
    Returns:
    rho -- the estimated autocorrelation coefficient
    results -- the fitted OLS model
    """
    # Initialize rho at 0
    rho = 0

    # Initialize delta at a value larger than the convergence criterion
    delta = np.inf

    # Loop until convergence
    i=0
    while delta > 1e-4:

        # Transform y and X
        y_star = y[1:] - rho * y[:-1]
        X_star = X[1:] - rho * X[:-1]

        # Fit model
        # print(y_star.shape, X_star.shape)
        model = sm.OLS(y_star, X_star)
        results = model.fit()
        
        if ax is not None:
            ax.plot(X, results.predict(X), f'{c}-', alpha=0.1)

        # Compute residuals
        residuals = results.resid

        # Update rho
        rho_new = np.sum(residuals[:-1] * residuals[1:]) / np.sum(residuals[:-1] ** 2)

        # Compute difference
        delta = np.abs(rho_new - rho)

        # Update rho
        rho_old = rho
        rho = rho_new
        
        # convergence failure
        if i > 90:
            print(rho_old, rho_new, delta)
        if i > n_it:
            print('C-O estimation failed to converge')
            return rho, results
            
        i+=1

    return rho, results
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

# Step 1: Fit model for each subject and get parameter estimates and standard errors
def fit_subjects(subjects, y, X):
    subject_params = []
    for subject in subjects:
        y_subject = y[subject] # assuming y is a dict with subject data
        X_subject = X[subject] # assuming X is a dict with subject data

        _, result = cochrane_orcutt(y_subject, X_subject)  # use the function you created earlier
        params = result.params
        se = result.bse
        subject_params.append((params, se))
    return subject_params

# Step 2: Bootstrap estimates for each model parameter on every subject
def bootstrap_estimates(subject_params):
    bootstrapped_params = []
    for params, se in subject_params:
        bootstrapped_params_subject = norm.rvs(loc=params, scale=se)
        bootstrapped_params.append(bootstrapped_params_subject)
    return bootstrapped_params

# Step 3: Find the mean of the bootstrap estimates across individuals
def mean_bootstrap_estimates(bootstrapped_params):
    return np.mean(bootstrapped_params, axis=0)

# Step 4: Predict population level curve
def predict_population(bootstrapped_mean_params, X):
    return np.dot(X, bootstrapped_mean_params)

# Step 5 and 6: Bootstrap population curves and calculate average response and standard deviation at each time point
def bootstrap_population(subjects, y, X, B=1000):
    bootstrapped_pop_curves = []
    for _ in range(B):
        subject_params = fit_subjects(subjects, y, X)
        bootstrapped_params = bootstrap_estimates(subject_params)
        bootstrapped_mean_params = mean_bootstrap_estimates(bootstrapped_params)
        population_prediction = predict_population(bootstrapped_mean_params, X)
        bootstrapped_pop_curves.append(population_prediction)
    
    # Convert list of arrays to 2D array for easier calculations
    bootstrapped_pop_curves = np.array(bootstrapped_pop_curves)
    
    # Mean response at each time point
    mean_responses = np.mean(bootstrapped_pop_curves, axis=0)
    
    # Standard deviation at each time point
    std_devs = np.std(bootstrapped_pop_curves, axis=0)
    
    return mean_responses, std_devs
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg

# Simulate some data
np.random.seed(0)
n = 100
X = np.random.normal(size=(n, 1))
ar_param = 0.7
errors = np.random.normal(size=n)
errors[1:] += ar_param * errors[:-1]  # This creates autocorrelation in the errors
y = 1 + X[:, 0] + errors  # y = 1 + X + error term with autocorrelation

# Fit the model
model = AutoReg(y, lags=1, exog=X, old_names=False)
result = model.fit()

ax=plt.figure().add_subplot()
ax.plot(X, y, 'b.')
ax.plot(X, errors, 'r.')
ax.plot(X[1:], result.fittedvalues, 'g-')
plt.show()

# Print the estimated parameters and standard errors
print("Estimated parameters:")
print(result.params)
print("Standard errors:")
print(result.bse)
```
